```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=data, x='Exited')
plt.title('Churn Distribution')
plt.show()
```

```python
sns.countplot(data=data , x='Gender', hue='Exited')
plt.title('Churn Distribution by Gender')
plt.show()
```

```python
sns.countplot(data=data, x='Geography', hue='Exited')
plt.title('Churn Distribution by Geography')
plt.show()
```

```python
sns.countplot(data=data,x='IsActiveMember',hue='Exited')
plt.title('Churn Distribution by IsActiveMember')
plt.show()
```

```python
# Credit Score Distribution for Churned and Retained Customers
sns.histplot(data=data, x='CreditScore', hue='Exited', kde=True, palette='infer')
plt.title('Credit Score Distribution (Exited vs. Retained)')
plt.show()
```

```python
# Age Distribution for Churned and Retained Customers
sns.histplot(data=data, x='Age', hue='Exited', kde=True, palette='inferno')
plt.title('Age Distribution (Exited vs. Retained)')
plt.show()
```

```python
# Tenure vs. Balance (Colored by Churn)
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Tenure', y='Balance', hue='Exited')
plt.title('Tenure vs. Balance (Colored by Churn)')
plt.show()
```

```python
# Churn rate by number of products
sns.countplot(data=data, x='NumOfProducts', hue='Exited', palette='plasma')
plt.title('Churn Rate by Number of Products')
plt.show()
```

```python
# Balance Distribution by Churn
sns.boxplot(data=data, x='Exited', y='Balance', palette='inferno')
plt.title('Balance Distribution by Churn')
plt.show()
```

```python
# estimated salary distribution by churn
sns.boxplot(data=data, x='Exited', y='EstimatedSalary', palette='inferno')
plt.title('Estimated Salary Distribution by Churn')
plt.show()
```

```python
# correlation heatmap
numeric_df = data[['CreditScore', 'Age', 'Tenure', 'Balance',
                 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
                 'EstimatedSalary', 'Exited']]

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='inferno', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
