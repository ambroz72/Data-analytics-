```python
# which location has highest & lowest sales
#represent it in a bar chart
data.head(5)
```

```python
location_list = data.groupby('Location')
```

```python
location = [x for x,y in location_list]
location
```

```python
data.groupby('Location').sum()['Total']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(location,data.groupby('Location').sum()['Total'])
plt.show()
```

```python
plt.pie(data.groupby('Location').sum()['Total'],labels=location,autopct = '%1.1f%%')
plt.show()
```

```python
# which locations has more female customers and which location has more male customers
loc_sales = data.groupby(['Gender','Location']).count()['Invoice ID']
loc_sales.unstack(level=0)
```

```python
us_sales = loc_sales.unstack(level=0)
us_sales.plot(kind='bar')
```
